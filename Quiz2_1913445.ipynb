```python
def displayImage(image):
    if len(image.shape)==3:
        color_coverted = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
        plt.imshow(color_coverted)
        plt.show()
        
    else:
        plt.imshow(image, cmap="gray")
        plt.show()
```

```python
img = cv2.imread(path+'lego1.png')
grayImage =cv2.cvtColor(img,cv2.COLOR_RGB2GRAY) #This will convert the RGB image to GRAY image
```

```python
displayImage(grayImage) #This will display the GRAY image
```

```python
blurred = cv2.GaussianBlur(grayImage, (7,7),25) #This will perform the Gaussian Blur
edged = cv2.Canny(blurred, 10, 45)  #This will perform the canny edge detection
```

```python
displayImage(edged) #This will show the canny edge output
```

```python
# define a (3, 3) structuring element
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3)) # This willperform morphological transforms
```

```python
# Apply the dilation operation to the edged image
dilate = cv2.dilate(edged, kernel, iterations=5) #This will perform dilation with 5 iterations
```

```python
(cnt,heirarchy)=cv2.findContours(dilate.copy(), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE) #This will retrieves contours from image
rgb=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
cv2.drawContours(rgb, cnt, -1, (0,255,0),2) #This will start to draw counter outlines

plt.imshow(rgb)
```
